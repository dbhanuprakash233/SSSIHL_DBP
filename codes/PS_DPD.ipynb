```python
"""
Plot the PMFs of six discrete distributions:
Discrete Uniform, Binomial, Poisson, Geometric, Negative Binomial, Hypergeometric.

Requires: numpy, scipy, matplotlib
    pip install numpy scipy matplotlib
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# ---- colors ----
BAR_COLOR = "#1F4E5F"     # bars
MEAN_COLOR = "#C97B2E"    # mean line

# ---- parameters for each distribution (edit these freely) ----
UNIFORM_A, UNIFORM_B = 1, 8
BINOM_N, BINOM_P = 12, 0.4
POISSON_LAMBDA = 4
GEOM_P = 0.3
NBINOM_R, NBINOM_P = 5, 0.4
HYPER_N, HYPER_K, HYPER_n = 40, 15, 10  # population, successes in population, sample size


def plot_pmf(ax, x, pmf, mean, title, xlabel):
    """Draw a PMF bar chart with a dashed mean line on the given axis."""
    ax.bar(x, pmf, color=BAR_COLOR, width=0.6)
    ax.axvline(mean, color=MEAN_COLOR, linestyle="--", linewidth=1.5, label="Mean")
    ax.set_title(title, fontsize=11, fontweight="bold")
    ax.set_xlabel(xlabel)
    ax.set_ylabel("P(X = k)")
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.legend(frameon=False, fontsize=8)


def main():
    fig, axes = plt.subplots(2, 3, figsize=(15, 8))

    # 1. Discrete Uniform ---------------------------------------------------
    x = np.arange(UNIFORM_A, UNIFORM_B + 1)
    pmf = np.full_like(x, 1 / len(x), dtype=float)
    mean = (UNIFORM_A + UNIFORM_B) / 2
    plot_pmf(axes[0, 0], x, pmf, mean,
              f"Discrete Uniform (a={UNIFORM_A}, b={UNIFORM_B})", "k")

    # 2. Binomial -------------------------------------------------------------
    x = np.arange(0, BINOM_N + 1)
    pmf = stats.binom.pmf(x, BINOM_N, BINOM_P)
    mean = BINOM_N * BINOM_P
    plot_pmf(axes[0, 1], x, pmf, mean,
              f"Binomial (n={BINOM_N}, p={BINOM_P})", "k")

    # 3. Poisson ---------------------------------------------------------------
    x = np.arange(0, POISSON_LAMBDA * 4)
    pmf = stats.poisson.pmf(x, POISSON_LAMBDA)
    mean = POISSON_LAMBDA
    plot_pmf(axes[0, 2], x, pmf, mean,
              f"Poisson (λ={POISSON_LAMBDA})", "k")

    # 4. Geometric (support starts at k = 1) -----------------------------------
    x = np.arange(1, int(5 / GEOM_P))
    pmf = stats.geom.pmf(x, GEOM_P)
    mean = 1 / GEOM_P
    plot_pmf(axes[1, 0], x, pmf, mean,
              f"Geometric (p={GEOM_P})", "k (trial of first success)")

    # 5. Negative Binomial (k = number of failures before r-th success) -------
    x = np.arange(0, int(NBINOM_R * 4 / NBINOM_P))
    pmf = stats.nbinom.pmf(x, NBINOM_R, NBINOM_P)
    mean = NBINOM_R * (1 - NBINOM_P) / NBINOM_P
    plot_pmf(axes[1, 1], x, pmf, mean,
              f"Negative Binomial (r={NBINOM_R}, p={NBINOM_P})", "k (number of failures)")

    # 6. Hypergeometric ---------------------------------------------------------
    lo = max(0, HYPER_n - (HYPER_N - HYPER_K))
    hi = min(HYPER_n, HYPER_K)
    x = np.arange(lo, hi + 1)
    pmf = stats.hypergeom.pmf(x, HYPER_N, HYPER_K, HYPER_n)
    mean = HYPER_n * HYPER_K / HYPER_N
    plot_pmf(axes[1, 2], x, pmf, mean,
              f"Hypergeometric (N={HYPER_N}, K={HYPER_K}, n={HYPER_n})", "k")

    fig.suptitle("PMFs of Six Discrete Distributions", fontsize=16, fontweight="bold")
    fig.tight_layout(rect=[0, 0, 1, 0.96])
    fig.savefig("discrete_distributions.png", dpi=200, bbox_inches="tight")
    plt.show()


if __name__ == "__main__":
    main()
```
